# QuTiP



[QuTiP (Quantum Toolbox in Python)](https://qutip.org) is a multipurpose package for numerical simulation of a variety of quantum systems - in particular qubits, harmonic oscillators, etc. It is particularly targeted at solving dynamics of open quantum systems, but can just as well be used for e.g. simple quantum optics or quantum circuits.



QuTiP is one of the oldest quantum-related Python packages and is being heavily used and [cited](https://scholar.google.com/citations?view_op=view_citation&hl=en&user=mpzOQQEAAAAJ&citation_for_view=mpzOQQEAAAAJ:LkGwnXOMwfcC):
